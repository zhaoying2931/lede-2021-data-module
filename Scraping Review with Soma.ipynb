```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get("http://www.bbc.com")
doc = BeautifulSoup(response.text, 'html.parser')
```

```python
doc.find_all('h3', class_='media__title')
```

```python
doc.select('.media__title')
```

```python
titles = doc.select('.media__title')
for title in titles:
    print(title.text.strip())
```

```python
tags = doc.select('.media__tag')
for tag in tags:
    print(tag.text.strip())
```

```python
import pandas as pd
```

```python
pd.DataFrame({
    'title':titles,
    'tag':tags
})
```

```python
summaries = doc.select('.media__summary')
len(summaries)
```

```python
stories = doc.select('.media-list__item')
stories
```

```python
rows=[]
for story in stories:
    print('-------')
    row = {}
    row['title']=story.select_one('h3').text.strip()
    try:
        row['herf'] = story.select_one('.media__link, .reel__link')['href']
    except:
        print("Couldn't find a link")
    try:
        row['tag'] = story.select_one('.media__summary').text.strip()
    except:
        print("Couldn't find a summary")
    print(row)
    rows.append(row)
```

```python
rows
```

```python
df = pd.DataFrame(rows)
```
